```python
import json
from glob import glob
import os

def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
        id, posts = data['timeline_id'], data['posts']

    return id, posts

input_dir = 'tasks12/'
ids, posts = [], []

for file in glob(os.path.join(input_dir, '*.json')):
    # print(f"Loading {file}...")
    id, post = load_json(file)
    ids.append(id)
    posts.append(post)

print(f"Loaded {len(ids)} records from {len(posts)}.")


```

```python
from networkx import adamic_adar_index


switch = 0 
esc = 0
p_cnt = 0
ada = 0
malada = 0 
sub_cat = []
sub_ds = {
    "ada":[],
    "malada":[]
}

TAXONOMY = {
    'A': {
        'adaptive': [],
        'maladaptive': []
    },
    'B-S': {
        'adaptive': [],
        'maladaptive': []
    },
    'B-O': {
        'adaptive': [],
        'maladaptive': []
    },
    'C-S': {
        'adaptive': [],
        'maladaptive': []
    },
    'C-O': {
        'adaptive': [],
        'maladaptive': []
    },
    'D': {
        'adaptive': [],
        'maladaptive': []
    }
}
for post in posts:
    for p in post:
        p_cnt += 1
        if p['Switch'] == "S":
            switch += 1
        if p['Escalation'] == "E":
            esc += 1
        if p['evidence']["adaptive-state"]:
            sub_c = list(p['evidence']["adaptive-state"].keys())
            sub_ds['ada'].extend(sub_c)
            for c in sub_c:
                # print(c)
                if c == 'Presence':
                    continue
                TAXONOMY[c]['adaptive'].append(p['evidence']["adaptive-state"][c]['Category'])
                TAXONOMY[c]['adaptive'] = list(set(TAXONOMY[c]['adaptive']))
            ada += 1
        if p['evidence']["maladaptive-state"]:
            sub_c = list(p['evidence']["maladaptive-state"].keys())
            sub_ds['malada'].extend(sub_c)
            for c in sub_c:
                # print(c)
                if c == 'Presence':
                    continue
                TAXONOMY[c]['maladaptive'].append(p['evidence']["maladaptive-state"][c]['Category'])
                TAXONOMY[c]['maladaptive'] = list(set(TAXONOMY[c]['maladaptive']))
            malada += 1
        

sub_cat = list(filter(lambda x: x not in {'Presence'}, sub_cat))

from collections import Counter
for k, v in sub_ds.items():
    print(k, Counter(v))


# print(p_cnt, switch, esc, ada, malada, sub_ds)
```
