```python
import json
import re
```

```python
with open('test_predictions_ensemble.json', 'r') as f:
    results = json.load(f)
```

```python
final_results = []
for res in results:
    output = {}
    output['timeline_id'] = res['timeline_id']
    output['post_id'] = res['post_id']
    # for _, row in results:
    pred = res['prediction']
    # print("\n=== ABCD Element Presence ===")
    output['adaptive-state'] = {}
    output['maladaptive-state'] = {}
    # Adaptive
    if pred and pred.get('adaptive-state'):
        if pred['adaptive-state'].get('Presence'):
            # print(data)
            output['adaptive-state']['Presence'] = pred['adaptive-state']['Presence']

        for dim in ['A', 'B-S', 'B-O', 'C-S', 'C-O', 'D']:
            if dim in pred['adaptive-state'] and pred['adaptive-state'][dim].get('Category'):
                cat = pred['adaptive-state'][dim]['Category']
                num = re.findall(r'\d+', cat)
                output['adaptive-state'][dim] = {'subelement': int(num[0])}
    else:
        output['adaptive-state']['Presence'] = 1
    
    # Maladaptive
    if pred and pred.get('maladaptive-state'):
        if pred['maladaptive-state'].get('Presence'):
            # print(data)
            output['maladaptive-state']['Presence'] = pred['maladaptive-state']['Presence']
        for dim in ['A', 'B-S', 'B-O', 'C-S', 'C-O', 'D']:
            if dim in pred['maladaptive-state'] and pred['maladaptive-state'][dim].get('subelement'):
                cat = pred['maladaptive-state'][dim]['subelement']
                num = re.findall(r'\d+', cat)
                output['maladaptive-state'][dim] = {'subelement': int(num[0])}
    else:
        output['maladaptive-state']['Presence'] = 1

                

        # print(f"Timeline ID: {tid}, Post ID: {pid}")
        # print("Adaptive Categories:")
        # print(adaptive)
        # print("Maladaptive Categories:")
        # print(maladaptive)
    final_results.append(output)

    # print("\nAdaptive:")
    # for dim in ['A', 'B-S', 'B-O', 'C-S', 'C-O', 'D']:
    #     print(f"  {dim}: {adaptive_counts[dim]}")
    
    # print("\nMaladaptive:")
    # for dim in ['A', 'B-S', 'B-O', 'C-S', 'C-O', 'D']:
    #     print(f"  {dim}: {maladaptive_counts[dim]}")

```

```python
len(final_results)
```

```python
with open('task1_pred.json', 'w') as f:
    json.dump(final_results, f, indent=2)
print("✅ Predictions saved to task1_pred.json\n")
```

```python
with open('task2_test_predictions_3way_qwen.json', 'r') as f:
    t2_results = json.load(f)
```

```python
final_results_t2 = []
for res2 in t2_results:
    output = {}
    output['timeline_id'] = res2['timeline_id']
    output['post_id'] = res2['post_id']
    pred2 = res2['prediction']
    print(pred2)
    if pred2 == "Neither":
        output['Switch'] = "0"
        output['Escalation'] = "0"
    elif pred2 == "Switch":
        output['Switch'] = "S"
        output['Escalation'] = "0"
    elif pred2 == "Escalation":
        output['Switch'] = "0"
        output['Escalation'] = "E"
    else:
        output['Switch'] = "S"
        output['Escalation'] = "E"

    final_results_t2.append(output)
```

```python
len(final_results_t2)
```

```python
with open('task2_pred.json', 'w') as f:
    json.dump(final_results_t2, f, indent=2)
print("✅ Predictions saved to task2_pred.json\n")
```
